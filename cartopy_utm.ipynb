```python
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import os
```

```python
boundaries = gpd.read_file('D:/Data/GIS/country/Country.shp')
print "crs: " + boundaries.crs['init']

crs = ccrs.PlateCarree()
fig = plt.figure()
ax = fig.add_subplot(111, facecolor='None', projection=crs)
boundaries.plot(ax=ax)
plt.show()
```

```python
boundaries_utm = gpd.read_file('D:/Data/GIS/country/Country_utm.shp')
print "crs: " + boundaries_utm.crs['init']

crs = ccrs.UTM(51)
fig = plt.figure()
ax = fig.add_subplot(111, facecolor='None', projection=crs)
boundaries.plot(ax=ax)
plt.show()
```
